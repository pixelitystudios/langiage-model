```python
import json
import re
```

```python
with open('./data/messages.json') as f:
    messages = json.load(f)
```

```python
messages.reverse()
```

```python
len(messages)
```

```python
data = []
```

```python
for idx, val in enumerate(messages):
    if(idx > 0):
        if(messages[idx - 1]['from'] == val['from']):
            data[len(data) - 1]['message'] = data[len(data) - 1]['message'] + ', ' + val['message']
        else:
            data.append(val)
    else:
        data.append(val)
```

```python
data = data[1:]
```

```python
len(data)
```

```python
questions = []
answers = []
limit = int(len(data) / 2)
```

```python
data
```

```python
qi = list(filter((lambda x: x%2==0),range(0,limit)))
ai = list(filter((lambda x: x%2!=0),range(0,limit)))
```

```python
for idx, val in enumerate(qi):
    questions.append(data[val]['message'].replace("Abraham Huerta", "jaja").replace("Viridiana Gzz", " jaja").replace("Has enviado una foto.", "").replace("Mi huevito de raptor ha enviado un sticker.", "").replace("Mi huevito de raptor ha enviado un GIF desde Tenor GIF Keyboard.", "").strip())
```

```python
for idx, val in enumerate(ai):
    answers.append(data[val]['message'].replace("Abraham Huerta", "jaja").replace("Viridiana Gzz", " jaja").replace("Has enviado una foto.", "").replace("Mi huevito de raptor ha enviado un sticker.", "").replace("Mi huevito de raptor ha enviado un GIF desde Tenor GIF Keyboard.", "").strip())
```

```python
print(questions[20])
print(answers[20])
```

```python
questions = "\n".join(questions)
```

```python
answers = "\n".join(answers)
```

```python
questions = re.sub(r'^https?:\/\/.*[\r\n]*', '', questions, flags=re.MULTILINE)
```

```python
questions
```

```python
with open("./data/Questions.txt", "w") as f: 
    f.write(questions) 
```

```python
with open("./data/Answers.txt", "w") as f: 
    f.write(answers) 
```
